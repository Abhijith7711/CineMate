```python
import pandas as pd
import pickle
```

```python
data  = pd.read_csv("Movies.csv")
```

```python
data.head()
```

```python
# to convert string field list into actual list
import ast

# Convert stringified lists into actual lists
data['genre'] = data['genre'].apply(lambda x: ', '.join(ast.literal_eval(x)) if x else '')
data['cast'] = data['cast'].apply(lambda x: ', '.join(ast.literal_eval(x)) if x else '')
data['director'] = data['director'].apply(lambda x: ', '.join(ast.literal_eval(x)) if x else '')
data['writer'] = data['writer'].apply(lambda x: ','.join(ast.literal_eval(x)) if x else '')
data['language'] = data['language'].apply(lambda x: ','.join(ast.literal_eval(x)) if x else '')
```

```python
#filling the missing descriptiom with a genre, director, cast
data['description'] = data.apply(lambda x: 
    x['genre'] + ' movie directed by ' + x['director'] + ' starring ' + x['cast'] 
    if pd.isna(x['description']) or x['description'].strip() == '' 
    else x['description'], axis=1)
```

```python
# to extract only year from released column
import re

# Function to extract year from date
def extract_year(date_str):
    match = re.search(r'\d{4}', str(date_str))  # Find 4-digit year
    return int(match.group()) if match else None  # Return year if found, else None

# Apply function to the 'released' column
data['released'] = data['released'].apply(extract_year)
```

```python
#filling missing values in the released column with median
data['released'] = data['released'].fillna(data['released'].median())

```
