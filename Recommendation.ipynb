```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
movie= pd.read_csv("moviedf.csv")
```

```python
movie
```

```python
movie.info()
```

```python
#features using for measuring similarity -

#description
#language
#writer
#director
#cast
#genre
```

```python
#combining features
features = movie['description']+ ' ' +movie['language']+' '+movie['writer'] + ' ' +movie['director'] +' '+ movie['cast']+' '+movie['genre']
```

```python
movie
```

```python
print(features)
```

```python
#tfif vectorization
tfidf = TfidfVectorizer(stop_words = 'english')
tfidf_matrix = tfidf.fit_transform(features)
```

```python
print(tfidf_matrix)
```

#### Cosine Similarity


```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim
```

```python
#save the cosine similaalrity cculation, tf-idf matrix,dataframe
pickle.dump(tfidf_matrix,open("tfidf_vectorizer.pkl","wb"))
pickle.dump(cosine_sim,open("cosine_similarity.pkl","wb"))
```

#### Recommendation Function

```python
#  Create Movie Recommendation Function
def recommend_movies(movie_name, num_recommendations=5):
    # Check if the movie exists
    if movie_name not in movie['name'].values:
        return f"Movie '{movie_name}'movie not found ."

    # Get the index of the movie
    idx = movie[movie['name'] == movie_name].index[0]

    # Get similarity scores for all movies
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort movies based on similarity score
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:num_recommendations+1]

    # Get the recommended movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return top recommended movie names
    return movie[['name','description','language','genre','cast']].iloc[movie_indices]

```

```python
#  Test the Recommendation System
movie_input = "3 Idiots"
print(recommend_movies(movie_input, 5))
```
